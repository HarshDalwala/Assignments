If the given string already ends with 'ing' then add 'ly' instead If the string length of the given
string is less than 3, leave it unchanged 

```python
string = input("Enter string: ")

if len(string) < 3:
    print(string)

elif string.endswith("ing"):
    print(string + "ly")

else:
    print(string + "ing")
```

6. Write a Python program to find the first appearance of the substring 'not' and 'poor' from a given string, if
'not' follows the 'poor', replace the whole 'not'...'poor' substring with 'good'.
Return the resulting string

```python
string = input("Enter string: ")

not_pos = string.find("not")
poor_pos = string.find("poor")

if not_pos != -1 and poor_pos != -1 and not_pos < poor_pos:
    result = string[:not_pos] + "good" + string[poor_pos + 4:]
    print(result)
else:
    print(string)
```

7. Program to find Greatest Common Divisor of two numbers.
For example, the GCD of 20 and 28 is 4 and GCD of 98 and 56 is 14.

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

while b:
    a, b = b, a % b

print("GCD =", a)
```

8. Write a Python program to check whether a list contains a sublist.

```python
main_list = [1, 2, 3, 4, 5]
sub_list = [3, 4]

found = False

for i in range(len(main_list) - len(sub_list) + 1):
    if main_list[i:i + len(sub_list)] == sub_list:
        found = True
        break

print("Sublist exists:", found)
```

13. Write a Python program to sort a dictionary (ascending /descending) by value.


```python
my_dict = {
    'A': 50,
    'B': 20,
    'C': 80,
    'D': 40
}

ascending = dict(sorted(my_dict.items(), key=lambda x: x[1]))

descending = dict(sorted(my_dict.items(),
                         key=lambda x: x[1],
                         reverse=True))

print("Ascending:", ascending)
print("Descending:", descending)
```

16. Counting the frequencies in a list using a dictionary in Python.
Input : [1, 1, 1, 5, 5, 3, 1, 3, 3, 1,4, 4, 4, 2, 2, 2, 2]
Expected output : 1 : 5 , 2 : 4 , 3 : 3 , 4 : 3 , 5 : 2

```python
numbers = [1,1,1,5,5,3,1,3,3,1,4,4,4,2,2,2,2]

freq = {}

for num in numbers:
    freq[num] = freq.get(num, 0) + 1

for key in sorted(freq):
    print(key, ":", freq[key])
```

17. Write a python program using function to find the sum of odd series and even series
Odd series: 12/ 1! + 32/ 3! + 52/ 5!+……n
Even series: 22/ 2! + 42/ 4! + 62/ 6!+……n

```python
import math

def odd_series(n):
    total = 0

    for i in range(1, n + 1, 2):
        total += (i ** 2) / math.factorial(i)

    return total


def even_series(n):
    total = 0

    for i in range(2, n + 1, 2):
        total += (i ** 2) / math.factorial(i)

    return total


n = int(input("Enter n: "))

print("Odd Series Sum =", odd_series(n))
print("Even Series Sum =", even_series(n))
```

18. Python Program to Find Factorial of Number Using Recursion


```python
def factorial(n):

    if n == 0 or n == 1:
        return 1

    return n * factorial(n - 1)


num = int(input("Enter number: "))

print("Factorial =", factorial(num))
```

```python
import random
import string


class User:

    def __init__(self, user_id, name, password):
        self.details = (user_id, name, password)

    def display(self):
        print("User Details:", self.details)


def generate_password(text):

    words = text.split()

    if len(words) < 2:
        raise ValueError(
            "Please enter at least two words."
        )

    selected = random.sample(
        words,
        min(2, len(words))
    )

    password = ""

    for word in selected:
        password += word.capitalize()

    password += str(random.randint(100, 999))

    password += random.choice("@#$%&*!")

    if len(password) < 8:
        password += "ABC123"

    return password


try:

    user_id = int(input("Enter User ID: "))
    name = input("Enter Name: ")
    text = input(
        "Enter some words for password generation: "
    )

    password = generate_password(text)

    user = User(
        user_id,
        name,
        password
    )

    print("Generated Password:", password)

    user.display()

except ValueError as e:
    print("Error:", e)

except Exception as e:
    print("Unexpected Error:", e)
```
