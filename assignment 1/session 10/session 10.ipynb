```python
# Create figure
plt.figure(figsize=(8,5))

# Draw line plot
plt.plot(days, steps)

# Add chart title
plt.title("Daily Steps Over Last 7 Days")

# Add X-axis label
plt.xlabel("Days")

# Add Y-axis label
plt.ylabel("Number of Steps")

# Display chart
plt.show()
```

```python
# Restaurant names
restaurants = [
    "Spice Villa",
    "Urban Cafe",
    "Pizza Hub",
    "Food Factory",
    "Royal Dine",
    "Green Bowl",
    "Burger Point",
    "BBQ Nation",
    "Tea Junction",
    "South Express"
]

# Zomato ratings (X-axis)
ratings = [4.5, 4.2, 3.8, 4.0, 4.7, 3.9, 4.1, 4.6, 3.7, 4.3]

# Average meal price in ₹ (Y-axis)
meal_price = [800, 600, 450, 500, 1200, 400, 550, 1000, 300, 700]

# Display data
print("Restaurant Data:\n")

for i in range(len(restaurants)):
    print(restaurants[i], "-", ratings[i], "-", meal_price[i])
```

```python
# Create figure
plt.figure(figsize=(8,5))

# Create scatter plot
plt.scatter(ratings, meal_price)

# Add chart title
plt.title("Zomato Ratings vs Average Meal Price")

# Add X-axis label
plt.xlabel("Zomato Rating")

# Add Y-axis label
plt.ylabel("Average Meal Price (₹)")

# Display chart
plt.show()
```

```python
# Food delivery platforms
platforms = ["Swiggy", "Zomato", "Domino's"]

# Number of orders placed last month
orders = [18, 25, 12]

# Display data
print("Order Data:\n")

for i in range(len(platforms)):
    print(platforms[i], "-", orders[i], "orders")
```

```python
# Create figure
plt.figure(figsize=(8,5))

# Create bar chart with different colors
plt.bar(platforms, orders,
        color=["orange", "red", "blue"],
        label="Total Orders")

# Add chart title
plt.title("Orders Placed Last Month")

# Add X-axis label
plt.xlabel("Platforms")

# Add Y-axis label
plt.ylabel("Number of Orders")

# Display legend
plt.legend()

# Display chart
plt.show()
```

```python
# Duration (in minutes) of last 20 Spotify listening sessions
session_duration = [
    20, 35, 40, 15, 50,
    60, 25, 30, 45, 55,
    70, 65, 80, 22, 38,
    42, 58, 33, 48, 27
]

# Display data
print("Spotify Listening Session Durations (Minutes):")
print(session_duration)
```

```python
# Import matplotlib library
import matplotlib.pyplot as plt
```

```python
# Days of the week
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

# Instagram screen time (minutes)
screen_time = [45, 60, 50, 70, 65, 90, 80]

# Number of posts liked
posts_liked = [20, 35, 28, 40, 38, 55, 48]

# Display data
print("Instagram Screen Time:")
print(screen_time)

print("\nPosts Liked:")
print(posts_liked)
```

```python
# Create figure with 2 subplots (1 row, 2 columns)
fig, ax = plt.subplots(1, 2, figsize=(12,5))

# -----------------------------
# First Subplot - Line Plot
# -----------------------------
ax[0].plot(days, screen_time,
           color="blue",
           marker="o",
           label="Screen Time")

ax[0].set_title("Instagram Screen Time")
ax[0].set_xlabel("Days")
ax[0].set_ylabel("Minutes")
ax[0].legend()

# -----------------------------
# Second Subplot - Bar Chart
# -----------------------------
ax[1].bar(days, posts_liked,
          color="orange",
          label="Posts Liked")

ax[1].set_title("Posts Liked Per Day")
ax[1].set_xlabel("Days")
ax[1].set_ylabel("Number of Likes")
ax[1].legend()

# Adjust spacing
plt.tight_layout()

# Save figure
plt.savefig("social_media_usage.png")

# Display figure
plt.show()

print("Figure saved successfully as 'social_media_usage.png'")
```
