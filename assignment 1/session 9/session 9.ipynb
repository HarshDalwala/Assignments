```python
# Import pandas library
import pandas as pd

# Create delivery timestamp dataset
data = {
    "Delivery_ID": [101, 102, 103],
    "Delivery_Time": [
        "2024-06-01 14:30",
        "2024-06-02 09:15",
        "2024-06-03 20:45"
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display original DataFrame
print("Original DataFrame:\n")
print(df)
```

```python
# Display data types before conversion
print("Data Types Before Conversion:\n")
print(df.dtypes)
```

```python
# Convert Delivery_Time column to datetime
df["Delivery_Time"] = pd.to_datetime(df["Delivery_Time"])

# Display updated DataFrame
print("Updated DataFrame:\n")
print(df)
```

```python
# Display data types after conversion
print("Data Types After Conversion:\n")
print(df.dtypes)
```

```python
# Import pandas library
import pandas as pd

# Load CSV file
df = pd.read_csv("flipkart_orders.csv")

# Display original dataset
print("Original Dataset:\n")
print(df)
```

```python
# Convert order_date column to datetime format
df["order_date"] = pd.to_datetime(df["order_date"])

# Display data types
print("Data Types:\n")
print(df.dtypes)
```

```python
# Extract year from order_date
df["Year"] = df["order_date"].dt.year

# Extract month from order_date
df["Month"] = df["order_date"].dt.month

# Extract weekday name from order_date
df["Weekday"] = df["order_date"].dt.day_name()

# Display updated DataFrame
print("Updated DataFrame:\n")
print(df)
```

```python
# Import pandas library
import pandas as pd

# Load CSV file
df = pd.read_csv("flipkart_orders.csv")

# Display original dataset
print("Original Dataset:\n")
print(df)
```

```python
# Convert order_date column to datetime
df["order_date"] = pd.to_datetime(df["order_date"])

# Display data types
print(df.dtypes)
```

```python
# Set order_date column as DataFrame index
df = df.set_index("order_date")

# Display DataFrame
print(df)
```

```python
# Calculate total number of orders each week
weekly_orders = df.resample("W").size()

# Display weekly order count
print("Weekly Total Orders:\n")
print(weekly_orders)
```

```python
# Import pandas library
import pandas as pd

# Create Instagram posts dataset
data = {
    "Post_ID": [101, 102, 103, 104, 105, 106],
    "Username": ["harsh", "raj", "priya", "amit", "neha", "rohan"],
    "posted_at": [
        "2024-06-01 08:00:00",
        "2024-06-01 12:30:00",
        "2024-06-02 15:45:00",
        "2024-06-03 18:20:00",
        "2024-06-04 21:10:00",
        "2024-06-05 05:15:00"
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display original DataFrame
print("Original DataFrame:\n")
print(df)
```

```python
# Convert posted_at column to datetime
df["posted_at"] = pd.to_datetime(df["posted_at"])

# Display data types
print(df.dtypes)
```

```python
# Localize timestamps to UTC
df["posted_at"] = df["posted_at"].dt.tz_localize("UTC")

# Display DataFrame
print("UTC Time:\n")
print(df)
```

```python
# Convert UTC timezone to Asia/Kolkata timezone
df["posted_at"] = df["posted_at"].dt.tz_convert("Asia/Kolkata")

# Display first 5 converted timestamps
print("First 5 Converted Times:\n")
print(df.head())
```

```python
# Import pandas library
import pandas as pd

# Load CSV file
df = pd.read_csv("flipkart_orders.csv")

# Display original dataset
print("Original Dataset:\n")
print(df)
```

```python
# Convert order_date column to datetime
df["order_date"] = pd.to_datetime(df["order_date"])

# Display data types
print(df.dtypes)
```

```python
# Create is_weekend column
# Saturday = 5, Sunday = 6
df["is_weekend"] = df["order_date"].dt.weekday >= 5

# Display updated DataFrame
print("Updated DataFrame:\n")
print(df)
```

```python
# Display only order_date and is_weekend columns
print(df[["order_date", "is_weekend"]])
```
