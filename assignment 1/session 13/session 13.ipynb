```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the built-in flights dataset
flights = sns.load_dataset("flights")

# Display the first 5 rows
flights.head()
```

```python
# Display dataset information
flights.info()
```

```python
# Convert the dataset into matrix format
flights_pivot = flights.pivot_table(
    values="passengers",
    index="month",
    columns="year",
    observed=False
)

# Display the pivot table
flights_pivot
```

```python
# Set figure size
plt.figure(figsize=(12, 7))

# Create heatmap
sns.heatmap(
    flights_pivot,
    annot=True,
    fmt=".0f",          # Display floats as whole numbers
    cmap="YlGnBu",
    linewidths=0.5
)

# Add title and labels
plt.title("Heatmap of Flight Passengers")
plt.xlabel("Year")
plt.ylabel("Month")

# Show the plot
plt.show()
```

```python
# Load the built-in fmri dataset
fmri = sns.load_dataset("fmri")

# Display the first 5 rows
fmri.head()
```

```python
# Display dataset information
fmri.info()
```

```python
# Create a relplot
sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",
    kind="line",
    height=5,
    aspect=1.5
)

# Add title
plt.title("Signal Changes Over Time for Different Event Types")

# Display the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the built-in titanic dataset
titanic = sns.load_dataset("titanic")

# Display first 5 rows
titanic.head()
```

```python
# Display dataset information
titanic.info()
```

```python
# Create catplot showing survival rate by passenger class
g = sns.catplot(
    data=titanic,
    x="class",
    y="survived",
    kind="bar",
    errorbar=("ci", 95),   # 95% Confidence Interval
    height=5,
    aspect=1.5
)

# Add title and labels
g.set_axis_labels("Passenger Class", "Survival Rate")
g.fig.suptitle("Survival Rate by Passenger Class", y=1.03)

# Display plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the built-in penguins dataset
penguins = sns.load_dataset("penguins")

# Display first 5 rows
penguins.head()
```

```python
# Display dataset information
penguins.info()
```

```python
# Remove rows with missing values in required columns
penguins_clean = penguins.dropna(
    subset=["bill_length_mm", "flipper_length_mm"]
)

# Display first 5 rows
penguins_clean.head()
```

```python
# Create jointplot with regression line
sns.jointplot(
    data=penguins_clean,
    x="bill_length_mm",
    y="flipper_length_mm",
    kind="reg",
    height=7
)

# Display plot
plt.show()
```
