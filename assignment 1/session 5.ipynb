```python
import sqlite3
import pandas as pd

# Create database
conn = sqlite3.connect("zomato.db")

# Create restaurants table
conn.execute("""
CREATE TABLE IF NOT EXISTS restaurants (
    id INTEGER,
    name TEXT,
    city TEXT,
    rating REAL
)
""")

# Insert records
conn.execute("DELETE FROM restaurants")

restaurants = [
    (1, "Dominos", "Surat", 4.2),
    (2, "Pizza Hut", "Ahmedabad", 4.1),
    (3, "Subway", "Vadodara", 4.0),
    (4, "McDonalds", "Surat", 4.3),
    (5, "Burger King", "Rajkot", 3.9)
]

conn.executemany(
    "INSERT INTO restaurants VALUES (?,?,?,?)",
    restaurants
)

conn.commit()

# Load table into DataFrame
df = pd.read_sql("SELECT * FROM restaurants", conn)

print(df.head())
```

```python
import pandas as pd
from pathlib import Path

orders_path = Path("orders.csv")
users_path = Path("users.csv")

orders = pd.read_csv(orders_path)
users = pd.read_csv(users_path)

merged_df = pd.merge(
    orders,
    users,
    on="user_id"
)

print(merged_df[["username", "amount"]])
```

```python
import pandas as pd

today_orders = pd.DataFrame({
    "order_id":[201,202],
    "item":["Mobile","Laptop"],
    "price":[15000,45000]
})

yesterday_orders = pd.DataFrame({
    "order_id":[199,200],
    "item":["Watch","Tablet"],
    "price":[3000,12000]
})

combined_df = pd.concat(
    [today_orders, yesterday_orders],
    ignore_index=True
)

print(combined_df)
```
