```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Generate random delivery times (50 orders)
np.random.seed(10)  # For same random output every time

delivery_time = np.random.randint(15, 61, 50)

# Create DataFrame
df = pd.DataFrame({
    "Delivery_Time": delivery_time
})

# Display first 10 records
print("First 10 Records:\n")
print(df.head(10))
```

```python
# Create figure
plt.figure(figsize=(8,5))

# Create histogram
sns.histplot(
    data=df,
    x="Delivery_Time",
    bins=8,
    color="skyblue",
    edgecolor="black"
)

# Add title
plt.title("Distribution of Zomato Delivery Times")

# Add X-axis label
plt.xlabel("Delivery Time (Minutes)")

# Add Y-axis label
plt.ylabel("Number of Orders")

# Display chart
plt.show()
```

```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set random seed (same output every time)
np.random.seed(10)

# IPL Teams
teams = [
    "CSK",
    "MI",
    "RCB",
    "KKR",
    "GT",
    "RR",
    "LSG",
    "DC"
]

# Create empty lists
team_list = []
runs_list = []

# Generate 15 match scores for each team
for team in teams:
    scores = np.random.randint(120, 231, 15)

    for score in scores:
        team_list.append(team)
        runs_list.append(score)

# Create DataFrame
df = pd.DataFrame({
    "Team": team_list,
    "Runs": runs_list
})

# Display first 10 rows
print("First 10 Records:\n")
print(df.head(10))
```

```python
# Apply Seaborn darkgrid theme
sns.set_theme(style="darkgrid")

# Create figure
plt.figure(figsize=(10,6))

# Create violin plot
sns.violinplot(
    data=df,
    x="Team",
    y="Runs",
    hue="Team",
    palette="Set2",
    legend=False
)

# Add title
plt.title("IPL Team Run Distribution")

# Add X-axis label
plt.xlabel("IPL Teams")

# Add Y-axis label
plt.ylabel("Runs")

# Display chart
plt.show()
```

```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create a list of 40 Spotify song genres
genres = [
    "Pop","Rock","Hip-Hop","Classical",
    "Pop","Rock","Hip-Hop","Pop",
    "Classical","Rock","Pop","Hip-Hop",
    "Rock","Pop","Classical","Hip-Hop",
    "Pop","Rock","Pop","Hip-Hop",
    "Classical","Rock","Pop","Hip-Hop",
    "Rock","Pop","Classical","Pop",
    "Hip-Hop","Rock","Pop","Classical",
    "Rock","Pop","Hip-Hop","Pop",
    "Rock","Classical","Pop","Hip-Hop"
]

# Create DataFrame
df = pd.DataFrame({
    "Genre": genres
})

# Display first 10 records
print("First 10 Records:\n")
print(df.head(10))
```

```python
# Apply Seaborn theme
sns.set_theme(style="whitegrid")

# Create figure
plt.figure(figsize=(8,5))

# Create countplot
sns.countplot(
    data=df,
    x="Genre",
    hue="Genre",
    palette="Set2",
    legend=False
)

# Add title
plt.title("Number of Spotify Songs by Genre")

# Add X-axis label
plt.xlabel("Genre")

# Add Y-axis label
plt.ylabel("Number of Songs")

# Display chart
plt.show()
```

```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Daily step counts for one week
steps = [5200, 6800, 7500, 8200, 9100, 10500, 12000]

# Create DataFrame
df = pd.DataFrame({
    "Daily_Steps": steps
})

# Display DataFrame
print("Daily Step Count Data:\n")
print(df)
```

```python
# Apply Seaborn whitegrid theme
sns.set_theme(style="whitegrid")

# Create figure
plt.figure(figsize=(8,5))

# Create KDE plot
sns.kdeplot(
    data=df,
    x="Daily_Steps",
    fill=True,
    color="green",
    linewidth=2
)

# Add title
plt.title("Distribution of Daily Step Counts")

# Add X-axis label
plt.xlabel("Daily Steps")

# Add Y-axis label
plt.ylabel("Density")

# Display chart
plt.show()
```
