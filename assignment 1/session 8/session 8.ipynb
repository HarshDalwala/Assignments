```python
# Import pandas library
import pandas as pd

# Load CSV file
df = pd.read_csv("zomato_orders.csv")

# Display the original dataset
print("Original Dataset:\n")
print(df)
```

```python
# Find duplicate orders based on Restaurant_Name and Order_Date
duplicates = df[df.duplicated(subset=["Restaurant_Name", "Order_Date"])]

# Display duplicate orders
print("Duplicate Orders:\n")
print(duplicates)
```

```python
# Print indices of duplicate rows
print("Duplicate Row Indices:")
print(duplicates.index.tolist())
```

```python
# Import pandas library
import pandas as pd

# Load CSV file
df = pd.read_csv("flipkart_reviews.csv")

# Display original dataset
print("Original Dataset:\n")
print(df)
```

```python
# Count how many times each review appears
review_count = df["Review_Text"].value_counts()

# Display review frequencies
print("Review Frequency:\n")
print(review_count)
```

```python
# Display top 3 most common reviews
top_reviews = df["Review_Text"].value_counts().head(3)

print("Top 3 Most Common Reviews:\n")
print(top_reviews)
```

```python
# Import pandas library
import pandas as pd

# Create Spotify playlist dataset
data = {
    "Playlist_Name": [
        "Top Hits",
        "Chill Vibes",
        "Workout Mix",
        "Top Hits",
        "Party Songs",
        "Workout Mix",
        "Focus Music"
    ],
    
    "Creator_Username": [
        "harsh123",
        "raj456",
        "priya789",
        "harsh123",
        "amit101",
        "priya789",
        "neha202"
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display original dataset
print("Original DataFrame:\n")
print(df)
```

```python
# Find duplicate rows
duplicates = df[df.duplicated()]

# Display duplicate rows
print("Duplicate Rows:\n")
print(duplicates)
```

```python
# Remove duplicate rows
clean_df = df.drop_duplicates()

# Display cleaned DataFrame
print("Cleaned DataFrame:\n")
print(clean_df)
```

```python
# Display number of rows before and after removing duplicates
print("Rows Before Removing Duplicates :", len(df))
print("Rows After Removing Duplicates  :", len(clean_df))
```

```python
# Import pandas library
import pandas as pd

# Create Instagram username dataset
data = {
    "Username": [
        "insta_queen",
        "insta-queen",
        "instaqueen",
        "tech_guru",
        "insta_queen",
        "food_blog",
        "insta-queen"
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display original DataFrame
print("Original DataFrame:\n")
print(df)
```

```python
# Replace different username variants with a standard username
df["Username"] = df["Username"].replace({
    "insta_queen": "instaqueen",
    "insta-queen": "instaqueen"
})

# Display updated DataFrame
print("Updated DataFrame:\n")
print(df)
```

```python
# Display unique usernames after replacement
print("Unique Usernames:\n")
print(df["Username"].unique())
```

```python
# Import pandas library
import pandas as pd

# Create Paytm payment status dataset
data = {
    "Customer_Name": [
        "Harsh",
        "Raj",
        "Priya",
        "Amit",
        "Neha",
        "Rohan",
        "Karan",
        "Sneha"
    ],
    
    "Payment_Status": [
        "Yes",
        " yes ",
        "Y",
        "No",
        " no ",
        "N",
        "YES",
        " n "
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display original DataFrame
print("Original DataFrame:\n")
print(df)
```

```python
# Remove extra spaces
df["Payment_Status"] = df["Payment_Status"].str.strip()

# Convert all text to lowercase
df["Payment_Status"] = df["Payment_Status"].str.lower()

# Convert values to 1 and 0 using map()
df["Payment_Status"] = df["Payment_Status"].map({
    "yes": 1,
    "y": 1,
    "no": 0,
    "n": 0
})

# Convert to integer
df["Payment_Status"] = df["Payment_Status"].astype(int)

# Display updated DataFrame
print("Updated DataFrame:\n")
print(df)
```

```python
# Display data types
print("Data Types:\n")
print(df.dtypes)
```
