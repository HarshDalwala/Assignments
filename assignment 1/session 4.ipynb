```python
import pandas as pd

# Read semicolon-separated CSV
df = pd.read_csv("paytm.csv", sep=';')

print("Dataset:")
print(df)

print("\nMissing Values Count:")
print(df.isnull().sum())

print("\nColumns Having Null Values:")

for col in df.columns:
    if df[col].isnull().sum() > 0:
        print(col)
```
