```python
# Import matplotlib library
import matplotlib.pyplot as plt
```

```python
# Days of the week
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

# Line chart data (Daily Sales)
sales = [120, 180, 150, 220, 200]

# Bar chart data (Orders)
orders = [12, 18, 15, 22, 20]

# Scatter plot data (Study Hours vs Marks)
study_hours = [1, 2, 3, 4, 5]
marks = [45, 55, 65, 80, 95]

# Pie chart data (Favorite Apps)
apps = ["Instagram", "YouTube", "WhatsApp", "Spotify"]
usage = [35, 30, 20, 15]
```

```python
# Create a figure with 2 rows and 2 columns
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

# ----------------------------------
# Line Plot
# ----------------------------------
ax[0, 0].plot(days, sales, marker="o")
ax[0, 0].set_title("Daily Sales")
ax[0, 0].set_xlabel("Days")
ax[0, 0].set_ylabel("Sales")

# ----------------------------------
# Bar Chart
# ----------------------------------
ax[0, 1].bar(days, orders, color="orange")
ax[0, 1].set_title("Orders")
ax[0, 1].set_xlabel("Days")
ax[0, 1].set_ylabel("Number of Orders")

# ----------------------------------
# Scatter Plot
# ----------------------------------
ax[1, 0].scatter(study_hours, marks, color="green")
ax[1, 0].set_title("Study Hours vs Marks")
ax[1, 0].set_xlabel("Study Hours")
ax[1, 0].set_ylabel("Marks")

# ----------------------------------
# Pie Chart
# ----------------------------------
ax[1, 1].pie(
    usage,
    labels=apps,
    autopct="%1.1f%%",
    startangle=90
)
ax[1, 1].set_title("Favorite Apps")

# Adjust spacing
plt.tight_layout()

# Display figure
plt.show()
```

```python
# Food delivery platforms
platforms = ["Zomato", "Swiggy", "Domino's"]

# Average delivery time (minutes)
delivery_time = [32, 28, 24]

# Display data
print("Average Delivery Time:\n")

for i in range(len(platforms)):
    print(platforms[i], "-", delivery_time[i], "minutes")
```

```python
# Create figure
plt.figure(figsize=(8,5))

# Create bar chart
plt.bar(
    platforms,
    delivery_time,
    color=["red", "orange", "blue"],   # Different colors
    edgecolor="black",                 # Border color
    linewidth=2,                       # Border thickness
    linestyle="--",                    # Dashed border
    label="Delivery Time"
)

# Add title
plt.title("Average Delivery Time Comparison")

# Add X-axis label
plt.xlabel("Food Delivery Platforms")

# Add Y-axis label
plt.ylabel("Average Delivery Time (Minutes)")

# Display legend
plt.legend()

# Display chart
plt.show()
```

```python
# Influencer names
influencers = ["Aman", "Priya", "Rohit", "Neha", "Karan"]

# Instagram followers
followers = [150000, 220000, 180000, 300000, 250000]

# Average daily posts
daily_posts = [2, 3, 1, 4, 2]

# Display dataset
print("Instagram Influencer Data:\n")

for i in range(len(influencers)):
    print(
        influencers[i],
        "- Followers:", followers[i],
        "- Daily Posts:", daily_posts[i]
    )
```

```python
# Create figure and first axis
fig, ax1 = plt.subplots(figsize=(8,5))

# ----------------------------------
# Left Y-axis (Followers)
# ----------------------------------
ax1.plot(
    influencers,
    followers,
    color="blue",
    marker="o",
    linewidth=2,
    label="Followers"
)

ax1.set_xlabel("Influencers")
ax1.set_ylabel("Followers", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

# ----------------------------------
# Right Y-axis (Daily Posts)
# ----------------------------------
ax2 = ax1.twinx()

ax2.plot(
    influencers,
    daily_posts,
    color="red",
    marker="s",
    linewidth=2,
    linestyle="--",
    label="Daily Posts"
)

ax2.set_ylabel("Average Daily Posts", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# Add title
plt.title("Instagram Followers vs Average Daily Posts")

# Display chart
plt.show()
```

```python
# Bollywood movie names
movies = [
    "Sikandar",
    "Housefull 5",
    "Raid 2",
    "Jaat",
    "Sky Force"
]

# IMDB ratings
ratings = [6.2, 5.8, 7.9, 6.8, 8.1]

# Tickets sold (in lakhs)
tickets_sold = [52, 38, 75, 45, 82]

# Display dataset
print("Bollywood Movie Data:\n")

for i in range(len(movies)):
    print(movies[i], "- Rating:", ratings[i], "- Tickets Sold:", tickets_sold[i], "Lakhs")
```

```python
# Create figure
plt.figure(figsize=(8,5))

# Create scatter plot
plt.scatter(
    ratings,
    tickets_sold,
    color="blue",
    s=100
)

# Add movie names above each point
for i in range(len(movies)):
    plt.annotate(
        movies[i],
        (ratings[i], tickets_sold[i]),
        xytext=(0,8),
        textcoords="offset points",
        ha="center"
    )

# Add title
plt.title("Tickets Sold vs IMDB Ratings")

# Add X-axis label
plt.xlabel("IMDB Rating")

# Add Y-axis label
plt.ylabel("Tickets Sold (Lakhs)")

# Display grid
plt.grid(True)

# Display chart
plt.show()
```

```python
# Months
months = [
    "Jan", "Feb", "Mar", "Apr",
    "May", "Jun", "Jul", "Aug",
    "Sep", "Oct", "Nov", "Dec"
]

# Monthly sales (in Crores)
sales = [120, 135, 150, 145, 160, 175, 180, 190, 250, 480, 300, 220]

# Display dataset
print("Flipkart Monthly Sales:\n")

for i in range(len(months)):
    print(months[i], "-", sales[i], "Crores")
```

```python
# Create figure and axis
fig, ax = plt.subplots(figsize=(10,5))

# Create line plot
ax.plot(
    months,
    sales,
    color="blue",
    marker="o",
    linewidth=2
)

# Find highest sales point
max_sales = max(sales)
max_index = sales.index(max_sales)

# Add annotation
ax.annotate(
    "Big Billion Days",
    xy=(months[max_index], sales[max_index]),
    xytext=(months[max_index], sales[max_index] + 40),
    arrowprops=dict(
        arrowstyle="->",
        color="red"
    ),
    fontsize=10,
    color="red",
    ha="center"
)

# Add title
ax.set_title("Flipkart Monthly Sales")

# Add X-axis label
ax.set_xlabel("Months")

# Add Y-axis label
ax.set_ylabel("Sales (Crores)")

# Display grid
ax.grid(True)

# Display chart
plt.show()
```
