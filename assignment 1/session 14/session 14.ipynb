```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Flipkart reviews dataset
reviews = pd.read_csv("flipkart_reviews.csv")

# Display first 5 rows
reviews.head()
```

```python
reviews.info()
```

```python
# Count the number of reviews for each rating
rating_count = reviews["rating"].value_counts().sort_index()

# Display the counts
rating_count
```

```python
# Set figure size
plt.figure(figsize=(8,5))

# Create bar plot
sns.countplot(
    data=reviews,
    x="rating",
    hue="rating",
    palette="viridis",
    legend=False
)

# Add title and labels
plt.title("Count of Reviews by Rating")
plt.xlabel("Rating (Stars)")
plt.ylabel("Number of Reviews")

# Show plot
plt.show()
```

```python
# Load the Zomato restaurant dataset
zomato = pd.read_csv("zomato_restaurants.csv")

# Display the first 5 rows
zomato.head()
```

```python
# Display dataset information
zomato.info()
```

```python
# Set figure size
plt.figure(figsize=(8,6))

# Create scatter plot
sns.scatterplot(
    data=zomato,
    x="average_cost_for_two",
    y="user_rating",
    s=100
)

# Add title and labels
plt.title("Relationship Between Cost and User Rating")
plt.xlabel("Average Cost for Two (₹)")
plt.ylabel("User Rating")

# Display the plot
plt.show()
```

```python
# Load the BookMyShow movie dataset
movies = pd.read_csv("bookmyshow_movies.csv")

# Display first 5 rows
movies.head()
```

```python
# Display dataset information
movies.info()
```

```python
# Find average box office collection for each genre
genre_avg = movies.groupby("genre")["box_office_collection"].mean()

# Display result
genre_avg
```

```python
# Sort by average box office collection in descending order
genre_avg_sorted = genre_avg.sort_values(ascending=False)

# Convert to DataFrame
genre_avg_sorted = genre_avg_sorted.reset_index()

# Rename column
genre_avg_sorted.columns = ["Genre", "Average Box Office Collection (Crores)"]

# Display sorted table
genre_avg_sorted
```
