```python
import pandas as pd

# Load CSV file
df = pd.read_csv("ipl_players.csv")

# Display original DataFrame
print("Original DataFrame:")
print(df)

# Shape before removing missing values
print("\nShape Before:", df.shape)

# Remove rows containing any missing value
new_df = df.dropna(axis=0, how='any')

# Shape after removing missing values
print("Shape After:", new_df.shape)

# Display cleaned DataFrame
print("\nDataFrame After dropna():")
print(new_df)
```

```python
import pandas as pd

# Load CSV file
df = pd.read_csv("ipl_players.csv")

# Display original DataFrame
print("Original DataFrame:")
print(df)

# Calculate mean of Runs column
mean_runs = df["Runs"].mean()

print("\nMean of Runs Column:", mean_runs)

# Replace missing values with mean
df["Runs"] = df["Runs"].fillna(mean_runs)

# Display updated Runs column
print("\nUpdated Runs Column:")
print(df["Runs"])
```

```python
import pandas as pd

# Create dataset
restaurants = {
    "Restaurant": [
        "Domino's",
        "Pizza Hut",
        "Subway",
        "McDonald's",
        "Burger King",
        "KFC"
    ],
    "City": [
        "Surat",
        "Ahmedabad",
        "Vadodara",
        "Rajkot",
        "Surat",
        "Ahmedabad"
    ],
    "Rating": [
        4.2,
        None,
        None,
        4.5,
        None,
        4.1
    ]
}

df = pd.DataFrame(restaurants)

print("Before Filling Missing Values:\n")
print(df)
```

```python
# Forward Fill
df["Rating"] = df["Rating"].ffill()

print("After Forward Fill:\n")
print(df)
```

```python
# Backward Fill (fills any remaining missing values)
df["Rating"] = df["Rating"].bfill()

print("After Backward Fill:\n")
print(df)
```

### Chosen Column: Rating

The missing values in the **Rating** column are most likely **MAR (Missing At Random)**.

This is because the missing ratings may depend on another observed variable, such as the restaurant's city or whether the restaurant is newly added to the dataset. Therefore, the missing values are related to other available data rather than being completely random.

**Missingness Mechanism:** **MAR (Missing At Random)**
