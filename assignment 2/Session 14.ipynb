```python
songs = [
    "Shape of You",
    "Levitating",
    "Blinding Lights",
    "Senorita",
    "Perfect"
]

with open("playlist.txt", "w") as file:
    for song in songs:
        file.write(song + "\n")

print("Songs written to playlist.txt")
```

```python
with open("playlist.txt", "r") as file:
    for song in file:
        print(song.strip().upper())
```

```python
import csv

with open("ipl_matches.csv", "r") as file:
    reader = csv.DictReader(file)

    for row in reader:
        print("Winner:", row["winner"])
```

```python
import json

with open("movies.json", "r") as file:
    movies = json.load(file)

for movie in movies:
    print(movie["title"], "-", movie["rating"])
```

```python
from pathlib import Path
import json

file_path = Path("my_fav_apps.json")

if not file_path.exists():
    apps = [
        {"name": "Instagram", "category": "Social Media"},
        {"name": "Zomato", "category": "Food Delivery"},
        {"name": "Paytm", "category": "Payments"}
    ]

    with open(file_path, "w") as file:
        json.dump(apps, file, indent=4)

    print("File created successfully.")
else:
    print("File already exists.")
```
