```python
import requests

session = requests.Session()

url = "https://jsonplaceholder.typicode.com/posts"

response1 = session.get(url)
print("First request status:", response1.status_code)

response2 = session.get(url)
print("Second request status:", response2.status_code)
```

```python
import requests

city = "Ahmedabad"
api_key = "YOUR_API_KEY_HERE"  # replace with real key

url = "https://api.openweathermap.org/data/2.5/weather"

params = {
    "q": city,
    "appid": api_key,
    "units": "metric"
}

response = requests.get(url, params=params)
data = response.json()

# Check if request was successful
if response.status_code == 200:
    print("Temperature in Ahmedabad:", data["main"]["temp"], "°C")
else:
    print("Error:", data.get("message", "Something went wrong"))
```

```python
import asyncio
import httpx

async def fetch(url):
    async with httpx.AsyncClient() as client:
        response = await client.get(url)
        return response.json()

async def main():
    url1 = "https://jsonplaceholder.typicode.com/posts/1"
    url2 = "https://jsonplaceholder.typicode.com/users/1"

    result1, result2 = await asyncio.gather(
        fetch(url1),
        fetch(url2)
    )

    print("Post Data:", result1)
    print("User Data:", result2)

await main()
```
