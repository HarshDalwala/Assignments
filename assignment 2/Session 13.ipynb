```python
def print_playlist_songs(songs):
    if len(songs) == 0:
        return

    print(songs[0])
    print_playlist_songs(songs[1:])

playlist = ["Shape of You", "Levitating", "Senorita", "Perfect"]

print_playlist_songs(playlist)
```

```python
def count_unread_messages(group):
    total = group["count"]

    for subgroup in group.get("subgroups", []):
        total += count_unread_messages(subgroup)

    return total

chat_groups = {
    "count": 10,
    "subgroups": [
        {
            "count": 5,
            "subgroups": []
        },
        {
            "count": 8,
            "subgroups": [
                {
                    "count": 3,
                    "subgroups": []
                }
            ]
        }
    ]
}

print("Total Unread Messages:", count_unread_messages(chat_groups))
```

```python
x = 'global'

def outer():
    x = 'outer'

    def inner():
        nonlocal x
        x = 'inner'

    inner()
    print('Inside outer:', x)

outer()
print('Outside:', x)
```

```python
def format_number_short(n):
    if n < 1000:
        return str(n)

    if n < 1000000:
        return str(round(n / 1000, 1)) + "K"

    return format_number_short(n / 1000) + "K"

print(format_number_short(500))
print(format_number_short(1500))
print(format_number_short(1200000))
```

```python
def format_number_short(n):
    if n < 1000:
        return str(n)
    elif n < 1000000:
        return f"{n/1000:.1f}K"
    else:
        return f"{n/1000000:.1f}M"

print(format_number_short(500))
print(format_number_short(1500))
print(format_number_short(1200000))
```
