```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def get_discounted_price(self):
        return self.price * 0.90   # 10% discount


class Electronics(Product):
    def get_discounted_price(self):
        return self.price * 0.80   # 20% discount


product = Product("T-Shirt", 1000)
electronic = Electronics("Laptop", 50000)

print(product.get_discounted_price())
print(electronic.get_discounted_price())
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def get_discounted_price(self):
        return self.price * 0.90


class Electronics(Product):
    def get_discounted_price(self):
        return self.price * 0.80


def show_final_price(item):
    print("Product:", item.name)
    print("Final Price:", item.get_discounted_price())


product = Product("T-Shirt", 1000)
electronic = Electronics("Laptop", 50000)

show_final_price(product)
show_final_price(electronic)
```

```python
class Ticket:
    def __init__(self, price):
        self.price = price

    def get_final_price(self):
        return self.price


class PremiumTicket(Ticket):
    def get_final_price(self):
        return super().get_final_price() + 50


normal_ticket = Ticket(200)
premium_ticket = PremiumTicket(200)

print("Normal Ticket Price:", normal_ticket.get_final_price())
print("Premium Ticket Price:", premium_ticket.get_final_price())
```
