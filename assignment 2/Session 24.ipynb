```python
import requests

url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1
}

try:
    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()

        for coin in data:
            print(f"{coin['name']} : ${coin['current_price']}")

    else:
        print("API Error: Unable to fetch data")

except requests.exceptions.RequestException:
    print("Network Error: Please check your internet connection")
```

```python
import requests

url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1
}

try:
    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()

        for coin in data:
            print("Name:", coin["name"])
            print("Price:", coin["current_price"])
            print("24h Change %:", coin["price_change_percentage_24h"])
            print("24h High:", coin["high_24h"])
            print("24h Low:", coin["low_24h"])
            print("-" * 30)

    else:
        print("API Error occurred")

except requests.exceptions.RequestException:
    print("Network Error")
```

```python
import requests
import csv

url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1
}

try:
    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()

        with open("crypto_prices.csv", "w", newline="") as file:
            writer = csv.writer(file)

            writer.writerow([
                "Name",
                "Price",
                "24h Change %",
                "24h High",
                "24h Low"
            ])

            for coin in data:
                writer.writerow([
                    coin["name"],
                    coin["current_price"],
                    coin["price_change_percentage_24h"],
                    coin["high_24h"],
                    coin["low_24h"]
                ])

        print("CSV file created successfully!")

    else:
        print("API Error")

except requests.exceptions.RequestException:
    print("Network Error")
```

```python
import requests

url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1
}

try:
    response = requests.get(url, params=params, timeout=10)

    response.raise_for_status()  # raises error if status != 200

    data = response.json()

    for coin in data:
        print(coin["name"], ":", coin["current_price"])

except requests.exceptions.HTTPError:
    print("HTTP Error: API returned an invalid response")

except requests.exceptions.ConnectionError:
    print("Connection Error: Check your internet")

except requests.exceptions.Timeout:
    print("Timeout Error: API took too long to respond")

except Exception as e:
    print("Something went wrong:", e)
```
