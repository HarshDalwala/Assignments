```python
import requests

data = {
    "title": "My Playlist",
    "body": "This is my first playlist created using Python",
    "userId": 1
}

response = requests.post(
    "https://jsonplaceholder.typicode.com/posts",
    json=data
)

print("Status Code:", response.status_code)
print("Response JSON:", response.json())
```

```python
import requests

title = input("Enter playlist name: ")
description = input("Enter playlist description: ")

data = {
    "title": title,
    "body": description,
    "userId": 1
}

response = requests.post(
    "https://jsonplaceholder.typicode.com/posts",
    json=data
)

result = response.json()

print("Playlist ID:", result.get("id"))
```

```python
import requests

data = {
    "name": "harsh",
    "job": "student"
}

response = requests.post(
    "https://reqres.in/api/users",
    json=data
)

result = response.json()

print("User ID:", result.get("id"))
print("Created At:", result.get("createdAt"))
```

```python
import requests
import csv

response = requests.get("https://jsonplaceholder.typicode.com/posts")
posts = response.json()[:5]

with open("posts.csv", "w", newline="") as file:
    writer = csv.writer(file)

    writer.writerow(["userId", "title"])

    for post in posts:
        writer.writerow([post["userId"], post["title"]])

print("CSV file created successfully!")
```

```python
import requests
import json

response = requests.get("https://jsonplaceholder.typicode.com/posts")
posts = response.json()[:5]

with open("posts.json", "w") as file:
    json.dump(posts, file, indent=4)

print("JSON file created successfully!")
```
