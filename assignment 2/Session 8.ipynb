```python
daily_steps = [6500, 8200, 9500, 11000, 12000, 9000, 10500]

day = 0

while day < len(daily_steps):
    if daily_steps[day] > 10000:
        print("First day crossing 10,000 steps:", day + 1)
        break
    day += 1
```

```python
def print_long_team_names(teams):
    for team in teams:
        if len(team) <= 6:
            continue
        print(team)

ipl_teams = [
    "CSK",
    "MI",
    "RCB",
    "Gujarat Titans",
    "Sunrisers Hyderabad"
]

print_long_team_names(ipl_teams)
```

```python
song_durations = [210, 185, 240, 195]

for position, duration in enumerate(song_durations, start=1):
    print(f"Song {position}: {duration} seconds")
```

```python
item_prices = [500, 700, 0, 600, 400, 300]

total = 0

for price in item_prices:
    if price == 0:
        continue  # Skip out-of-stock items

    total += price

    if total > 2000:
        break  # Stop when total exceeds ₹2000

print("Final Total:", total)
```
