```python
apps = ["Zomato", "Swiggy", "Domino's", "Pizza Hut", "Uber Eats"]

app_iterator = iter(apps)

print(next(app_iterator))
print(next(app_iterator))
print(next(app_iterator))
print(next(app_iterator))
print(next(app_iterator))
```

```python
def playlist_generator(songs):
    for song in songs:
        yield song

playlist = ["Shape of You", "Levitating", "Perfect", "Senorita"]

for song in playlist_generator(playlist):
    print(song)
```

```python
cart = ["Pizza", "Burger", "Fries", "Coke"]

for index, item in enumerate(cart, start=1):
    print(f"Item {index}: {item}")
```

```python
teams = ["Mumbai Indians", "Chennai Super Kings", "RCB"]
points = [18, 16, 14]

for team, point in zip(teams, points):
    print(f"Team: {team}, Points: {point}")
```

```python
def order_id_generator():
    order_id = 1001

    while True:
        yield order_id
        order_id += 1

orders = order_id_generator()

print(next(orders))
print(next(orders))
print(next(orders))
print(next(orders))
```
